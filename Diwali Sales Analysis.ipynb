# Exploratory Data Analysis

## Gender

```python
df.columns
```

```python
ax = sns.countplot(x = 'Gender', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.barplot(x = 'Gender', y = 'Amount', data = sales_gen)
```

## Age

```python
df.columns
```

```python
ax = sns.countplot(x = 'Age Group', hue = 'Gender', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.barplot(x = 'Age Group', y = 'Amount', data = sales_age)
```

## State

```python
sales_states = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)
sns.set(rc = {'figure.figsize':(20, 7)})
sns.barplot(x = 'State', y = 'Orders', data = sales_states)
```

```python
sales_states = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
sns.set(rc = {'figure.figsize':(20, 7)})
sns.barplot(x = 'State', y = 'Amount', data = sales_states)
```

## Marital status

```python
df.columns
```

```python
ax = sns.countplot(x = 'Marital_Status', data = df)
sns.set(rc = {'figure.figsize':(4,2)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.set(rc = {'figure.figsize':(6,5)})
sns.barplot(x = 'Marital_Status', y = 'Amount', hue = 'Gender', data = sales_state)
```

From above graphs we can see that most of the buyers are married (women) and they have high purchasing power

## Occupation

```python
df.columns
```

```python
sns.set(rc = {'figure.figsize':(20,5)})
ax = sns.countplot(x = 'Occupation', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_states = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
sns.set(rc = {'figure.figsize':(20, 5)})
sns.barplot(x = 'Occupation', y = 'Amount', data = sales_states)
```

From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector

## Product Category

```python
sns.set(rc = {'figure.figsize':(20,5)})
ax = sns.countplot(x = 'Product_Category', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_states = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
sns.set(rc = {'figure.figsize':(20, 5)})
sns.barplot(x = 'Product_Category', y = 'Amount', data = sales_states)
```

*From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category*

```python
sales_states = df.groupby(['Product_ID'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
sns.set(rc = {'figure.figsize':(20, 5)})
sns.barplot(x = 'Product_ID', y = 'Amount', data = sales_states)
```

# Conclusion :

*Married women age group 26-35 yrs from UP,  Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category*
